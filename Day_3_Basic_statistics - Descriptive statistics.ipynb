### **DESCRIPTIVE STATISTICS**
whatever the **statistical measures** that we can calculate from the sample data we call them **descriptive statistics**


statistical measures like mean,mode,meadian,std,var,range,min,max....

### **INFERENTIAL STATISTICS**
using the above calculations and apply some more additional theory and trying the estimate the entire population parameters is called **inferential statistics**
