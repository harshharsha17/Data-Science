 **Inter quartile range**:

 IQR = Q3 - Q1

 **Lower whisker length**:

 LWL = Q1 - (1.5 * IQR)

 **Upper whisker length**:

 UWL = Q3 + (1.5 * IQR)

 if any data points are falling either above to the upper whisker length and below to the lower lower whisker length are treated as **OUTLIERS**
