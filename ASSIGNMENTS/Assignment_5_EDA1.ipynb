Now lets see from the above lets divide the outliers into low outliers and high outliers columns 

Low outliers columns = Tendency, width, ASTV, AC, UC, LB, MLTV, MSTV

High outliers columns = DL, DS, DP, FM, ALTV, NSP

Now lets remove the outliers from low outliers columns and replace with whisker length in high outliers columns 

#### Removing outliers for low outlier columns 

```python
#before removing outliers shape of the dataset is 
df.shape
```

```python
low_outlier_columns=['Tendency', 'Width', 'ASTV', 'AC', 'UC', 'LB', 'MLTV', 'MSTV']
for col in low_outlier_columns:
    Q1=np.percentile(df[col],25)
    Q3=np.percentile(df[col],75)
    IQR=Q3-Q1
    LWL=Q1-(1.5*IQR)
    UWL=Q3+(1.5*IQR)
    outliers=df[(df[col]<LWL) | (df[col]>UWL)]
    print("Outliers in ",col,": ",len(outliers))
    print("Outlier percentage in ",col,": ",np.round((len(outliers)/len(df)*100),2))
    df=df[(df[col]>=LWL) & (df[col]<=UWL)]
```

```python
# after removing the outliers 
df.shape
```

#### Replacing  for high outlier columns

```python
#describe before replacing outliers 
df.describe()
```

```python
high_outlier_columns=['DL', 'DS', 'DP', 'FM', 'ALTV', 'NSP']
for col in high_outlier_columns:
    Q1=np.percentile(df[col],25)
    Q3=np.percentile(df[col],75)
    IQR=Q3-Q1
    LWL=Q1-(1.5*IQR)
    UWL=Q3+(1.5*IQR)
    outliers=df[(df[col]<LWL) | (df[col]>UWL)]
    print("Outliers in ",col,": ",len(outliers))
    print("Outlier percentage in ",col,": ",np.round((len(outliers)/len(df)*100),2))
    df[col]=np.where(df[col]<LWL,LWL,np.where(df[col]>UWL,UWL,df[col]))
```

```python
#describe after removing the outliers 
df.describe()
```
