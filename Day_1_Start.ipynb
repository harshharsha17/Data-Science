## **Data Science??**

 Data science is generation of actionable knowledge directly from huge amount of complex data.

 The goal of data science is to gain insights and knowlege from any type of data both structured and unstructured

 ## **Types of data**

 ### Structured data
 Data stored in rows and columns

 ex: education,hr,banking,hospitals,tours,ecommerce....

 ## Unstructured data
Data is not stored in rows and columns

ex: comments,images,videos,reviews,scanning reports....
